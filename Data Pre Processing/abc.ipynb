# Univariate Analysis
Bar charts : the simplest visualization chart that maps categories to numbers.

```python
iris_df['iris'].value_counts().plot.bar()
```

```python
iris_df['petal length'].plot.hist()
```

```python
iris_df2 = iris_df[(iris_df['petal length'] >= 1) & (iris_df['petal length'] <=1.5)]
iris_df2.count()
```
